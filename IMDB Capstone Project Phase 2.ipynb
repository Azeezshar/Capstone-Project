```python
# importing SQLite 3
import sqlite3

#connecting to the database, if there exsits no database, it will create one
db = sqlite3.connect("IMDb.db")

# Declare cursor
cursor = db.cursor()
```

```python
# Creating first Table IMDb_tab_1
cursor.execute(" CREATE TABLE IMDb_tab_1(serial_no INT, movie_name VARCHAR, director_1 VARCHAR, director_2 VARCHAR, director_3 VARCHAR, director_4 VARCHAR, director_5 VARCHAR, director_6 VARCHAR, duration VARCHAR, year_of_release VARCHAR, ratings VARCHAR, metascore VARCHAR)")
```

```python
# Creating second table IMDb_tab_2
cursor.execute(" CREATE TABLE IMDb_tab_2(Movie_Name varchar,Star_1 varchar,Star_2 varchar,Star_3 varchar,Star_4 varchar,Votes INT, Genre_1 text,Genre_2 text, Genre_3 text,Gross_Collection varchar,Certification varchar)")
```

```python
# Inserting the csv file 1 in the database and table
import pandas as pd

df = pd.read_csv('IMDb_1.csv', encoding="utf8")

try:
    df.to_sql('IMDb_tab_1', con=db, if_exists='append', index=False)
    total_record = len(df)
    print(total_record, "records inserted")
except Exception as e:
    print(f"Error inserting data: {e}")
```

```python
# checking the records inserted
results = cursor.execute("SELECT * FROM IMDb_tab_1")
for row in results:
    print(row)
```

#### Table 1

```python
# 1) Display all the details of movies created by directors Christopher and Matt Reeves.

query = '''
SELECT IMDb_tab_1.*, IMDb_tab_2.*
FROM IMDb_tab_1
JOIN IMDb_tab_2 ON IMDb_tab_1.movie_name = IMDb_tab_2.Movie_Name
WHERE IMDb_tab_1.director_1 IN (' Christopher Nolan ', ' Matt Reeves ')
   OR IMDb_tab_1.director_2 IN (' Christopher Nolan ', ' Matt Reeves ')
   OR IMDb_tab_1.director_3 IN (' Christopher Nolan ', ' Matt Reeves ')
   OR IMDb_tab_1.director_4 IN (' Christopher Nolan ', ' Matt Reeves ')
   OR IMDb_tab_1.director_5 IN (' Christopher Nolan ', ' Matt Reeves ')
   OR IMDb_tab_1.director_6 IN (' Christopher Nolan ', ' Matt Reeves ')
'''

results=cursor.execute(query)
result=results.fetchall()
for row in result:
    print(row)
```

```python
result_df = pd.read_sql_query(query, db)
result_df
```

```python
# 2) Display all the details of movies with a duration of 140 minutes to 190 minutes.

sql = '''SELECT IMDb_tab_1.*, IMDb_tab_2.* FROM IMDb_tab_1 JOIN IMDb_tab_2 
ON IMDb_tab_1.movie_name = IMDb_tab_2.Movie_Name
WHERE duration BETWEEN "140 min" and "190 min"'''

results=cursor.execute(sql)
result=results.fetchall()
for row in result:
    print(row)
```

```python
result_df = pd.read_sql_query(sql, db)
result_df
```

```python
# 3) Display all details of movies with ratings above 7 in ascending order.

sql = '''SELECT IMDb_tab_1.*, IMDb_tab_2.* FROM IMDb_tab_1 JOIN IMDb_tab_2 
ON IMDb_tab_1.movie_name = IMDb_tab_2.Movie_Name
WHERE ratings > 7.0 ORDER BY ratings asc'''

results=cursor.execute(sql)
result=results.fetchall()
for row in result:
    print(row)
```

```python
result_df = pd.read_sql_query(sql, db)
result_df
```

#### Table 2

```python
# 1) Display all movie names with star Arnold Schwarzenegger in ascending order.

sql = '''SELECT IMDb_tab_2.Movie_Name
FROM IMDb_tab_2
WHERE IMDb_tab_2.Star_1 = ' Arnold Schwarzenegger'
   OR IMDb_tab_2.Star_2 = ' Arnold Schwarzenegger'
   OR IMDb_tab_2.Star_3 = ' Arnold Schwarzenegger'
   OR IMDb_tab_2.Star_4 = ' Arnold Schwarzenegger'
ORDER BY IMDb_tab_2.Movie_Name ASC
'''
results=cursor.execute(sql)
result=results.fetchall()
for row in result:
    print(row)
```

```python
result_df = pd.read_sql_query(sql, db)
result_df
```

```python
# 2) Display all details of the movie with the highest number of votes.

sql = '''
SELECT IMDb_tab_1.*, IMDb_tab_2.*
FROM IMDb_tab_1
JOIN IMDb_tab_2 ON IMDb_tab_1.movie_name = IMDb_tab_2.Movie_Name
WHERE CAST(REPLACE(IMDb_tab_2.Votes, ',', '') AS INTEGER) = (
    SELECT MAX(CAST(REPLACE(Votes, ',', '') AS INTEGER))
    FROM IMDb_tab_2)
'''
results=cursor.execute(sql)
result=results.fetchall()
for row in result:
    print(row)
```

```python
result_df = pd.read_sql_query(sql, db)
result_df
```

```python
# 3) Display movie names with gross collections in descending order.

sql = '''
SELECT IMDb_tab_2.Movie_Name, REPLACE(REPLACE(IMDb_tab_2.Gross_Collection, '$', ''), 'M', '') AS Gross_Collection
FROM IMDb_tab_2
ORDER BY CAST(REPLACE(REPLACE(IMDb_tab_2.Gross_Collection, '$', ''), 'M', '') AS REAL) DESC
'''
results=cursor.execute(sql)
result=results.fetchall()
for row in result:
    print(row)
```

```python
result_df = pd.read_sql_query(sql, db)
result_df
```

```python
# 4) Display the gross collection of movies with the star Arnold.

sql='''
SELECT IMDb_tab_2.Gross_Collection, IMDb_tab_2.Movie_name
FROM IMDb_tab_2
WHERE IMDb_tab_2.Star_1 = ' Arnold Schwarzenegger'
   OR IMDb_tab_2.Star_2 = ' Arnold Schwarzenegger'
   OR IMDb_tab_2.Star_3 = ' Arnold Schwarzenegger'
   OR IMDb_tab_2.Star_4 = ' Arnold Schwarzenegger'
'''
results=cursor.execute(sql)
result=results.fetchall()
for row in result:
    print(row)
```

```python
# 5) Display all details of movies with comedy and action genres.

sql = '''
SELECT IMDb_tab_1.*, IMDb_tab_2.*
FROM IMDb_tab_1
JOIN IMDb_tab_2 ON IMDb_tab_1.movie_name = IMDb_tab_2.Movie_Name
WHERE (IMDb_tab_2.Genre_1 LIKE '%Comedy%' OR IMDb_tab_2.Genre_1 LIKE '%Action%')
  AND (IMDb_tab_2.Genre_2 LIKE '%Action%' OR IMDb_tab_2.Genre_2 LIKE '%Comedy%')
'''

results = cursor.execute(sql)
result = results.fetchall()
for row in result:
    print(row)
```
